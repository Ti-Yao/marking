```python
import pandas as pd
root_folder = 'drive/MyDrive/CARD0003/June_2024'
import re
# questions
file = f'{root_folder}/reference/Case-Based_Answer.xlsx'
marks = pd.read_excel(file)
marks = marks.set_index('ID')
columns = marks.columns.values
old_questions = [col for col in columns if len(re.findall(r"^Part", col)) > 0]
question_counter = 0
questions = []
for form_idx, question in enumerate(old_questions):
    if 'Part i)' in question:
        question_counter += 1
    questions.append(f"Form {form_idx + 2}: Q{question_counter}.{question.replace('Part','')}".replace('\n',''))

#responses
responses = marks[old_questions]
responses.columns = questions

# answers
df = pd.read_excel(f'{root_folder}/reference/CARD0003_Online_marking_template_MSc_2024_final.xlsx')
# df['Part'] = df.Part.apply(lambda x: int_to_roman(x))
# df['Question'] = 'Q' + df['Question'].astype(str) + ' Part ' + df['Part'] + ') ' + df['Question.1']
answers = pd.DataFrame(list(df['Model Answer'].iloc[:len(questions)].values),index = questions, columns = ['answer']).T

# # scores
scores = pd.DataFrame(list(df['Score'].iloc[:len(questions)].values),index = questions, columns = ['score']).T


formatted_reponses = pd.concat([responses, answers,scores])
formatted_reponses.index.Name = 'ID'
formatted_reponses.index = formatted_reponses.index.astype(str)
formatted_reponses.to_csv(f'{root_folder}/reference/formatted_reponses.csv')
```
