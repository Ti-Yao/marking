```python
import streamlit as st
import pandas as pd
import numpy as np
import os
import re

# Constants
root_folder = '.'
marking_groups = 'Marking Groups.xlsx'
marking_groups = pd.read_excel(f'{root_folder}/reference/{marking_groups}').iloc[2:]
marking_groups.columns = ['Q', 'mark', 'marker1', 'marker2']

# Streamlit UI for selecting marker initials
Marker_Initials = st.selectbox("Select Marker Initials", ["VM", "AC"])

# Filter questions based on selected marker initials
question_selection = list(marking_groups.loc[(marking_groups['marker1'] == Marker_Initials) | (marking_groups['marker2'] == Marker_Initials)].Q.values)
results_path = f'{root_folder}/results/scores_{Marker_Initials.upper()}.csv'

# Load formatted responses
formatted_responses = pd.read_csv(f'{root_folder}/reference/formatted_reponses.csv', index_col=0)
formatted_questions = questions = list(formatted_responses.columns)
formatted_questions = questions = [q for q in formatted_questions if int(q.split('Q')[-1].split('.')[0]) in question_selection]
formatted_responses = formatted_responses[questions]
candidates = formatted_responses.drop(['score', 'answer']).astype(str).index
scores = formatted_responses.loc['score']
answers = formatted_responses.loc['answer']
responses = formatted_responses.drop(['score', 'answer'])

# Initialize results file if it doesn't exist
if os.path.isfile(results_path):
    results = pd.read_csv(results_path)
else:
    results = pd.DataFrame()
    results['ID'] = candidates
    for q in questions:
        results[q] = '-'
    results.to_csv(results_path, index=False)

# Function to format strings
def format_string(string):
    if len(string.replace('\n', '')) > 90:
        formatted_string = re.sub(r'(.{1,90})\s', r'\1\n', string)
        return formatted_string
    else:
        return string

# Function to get unmarked questions
def get_unmarked(question):
    results = pd.read_csv(results_path)
    unmarked = list(results[question].values).count('-')
    return unmarked

# Function to save results
def save_results(drop_values, question):
    results = pd.read_csv(results_path)
    results[question] = drop_values
    results.to_csv(results_path, index=False)

# Streamlit UI
st.title("Marking Tool")

# Dropdown to jump to any question
question_index = st.selectbox("Jump to Any Question", range(len(questions)), format_func=lambda x: questions[x])

# Display current question
current_q = questions[question_index]
q = format_string(formatted_questions[question_index])
st.write(f"{Marker_Initials}. You are marking Questions: {', '.join(str(x) for x in question_selection)}")
st.write(f"**{format_string(q)}**")
st.write(f"**Answer:**\n{answers[question_index].replace('(2) ', '(2)\n')}")
st.write("---")

# Display candidate responses and dropdowns for marking
drop_values = []
for candidate in candidates:
    st.write(f"**Candidate {candidate}:**")
    response = responses.fillna('No Response').loc[str(candidate)][current_q].replace('\n', ' ').replace('\t', ' ')
    response = format_string(response)
    st.write(response)
    value = results[current_q].loc[str(candidate)]
    drop_value = st.selectbox(f"Mark for Candidate {candidate}", ['-'] + [str(val) for val in np.arange(0, float(scores[current_q]) + 0.5, 0.5)], index=0 if value == '-' else list(np.arange(0, float(scores[current_q]) + 0.5).index(float(value)) + 1)
    drop_values.append(drop_value)
    st.write("---")

# Save button
if st.button("Save Marks"):
    save_results(drop_values, current_q)
    st.success("Marks saved successfully!")

# Navigation buttons
col1, col2 = st.columns(2)
with col1:
    if st.button("Previous Question") and question_index > 0:
        question_index -= 1
with col2:
    if st.button("Next Question") and question_index < len(questions) - 1:
        question_index += 1

# Display unmarked questions
unmarked_questions = [q for q in questions if get_unmarked(q) > 0]
st.write(f"**Questions unmarked:** {len(unmarked_questions)}/{len(questions)}. Qs: {', '.join(unmarked_questions)}")
```
